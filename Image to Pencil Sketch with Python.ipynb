## Task (4) : Image to Pencil Sketch with Python :- 
We need to read the image in RBG format and then convert it to a grayscale
image. This will turn an image into a classic black and white photo. Then the
next thing to do is invert the grayscale image also called negative image, this
will be our inverted grayscale image. Inversion can be used to enhance details.
Then we can finally create the pencil sketch by mixing the grayscale image with
the inverted blurry image. This can be done by dividing the grayscale image by
the inverted blurry image. Since images are just arrays, we can easily do this
programmatically using the divide function from the cv2 library in Python.

## Import libaries


The only library we need for converting an image into a pencil sketch with Python is an OpenCV library in Python. It can be used by using the pip command; pip install opencv-python. But it is not imported by the same name. Let’s import it to get started with the task: import cv2




```python
import cv2
import matplotlib.pyplot as plt
```

### Step 0: Load and Show the Image

I will not display the image at every step, if we want to display the image at every step to see the changes in the image then we need to use two commands; cv2.imshow(“Title that we need to give”, Image) and then simply write cv2.waitKey(0). This will display the image.

Now the next thing to do is to read the image:

```python
# Load the image

img = cv2.imread("dog.jpg")
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

### Step1: Convert to Grey Image from original image

```python
# Convert to Grey Image

gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Show image

cv2.imshow("New Tree", gimg)
cv2.waitKey(0)
```

### Step2: Convert to Inverted Image from grey image

```python
# Invert Image
# Subtract the img from max value

inverted_image = 255 - gimg

# Show image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

### Step3: Convert to Blur Image from inverted image

```python
# Blur image
    
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2.imshow("Blurred", blurred)
cv2.waitKey(0)
```

### Step4: Convert to Inverted Blur Image from blur image

```python
# Invert Blurred Image

inverted_blurred = 255 - blurred
cv2.imshow("Inverted Blur",inverted_blurred )
cv2.waitKey(0)
```

### Step5 & Last Step: Divided the grey image by inverted blur image and get the Sketch image

```python
# Sketch Image

pencil_sketch = cv2.divide(gimg, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

### Displaying the original pic and sketch pic

```python
# Display original image & sketch image

cv2.imshow("original image", img)
cv2.imshow("pencil sketch", pencil_sketch)
cv2.waitKey(0)
```
